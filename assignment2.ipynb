```python
import pandas
import matplotlib
abalone_df = pandas.read_csv("abalone.data", names=['Sex', 'Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight', 'Viscera weight', 'Shell weight',
                                  'Rings'])
abalone_df


```

- For each categorical columns compute the mode.<br>

```python
abalone_df.mode()
```

- For each numerical column compute the mean (average), max, and min values.<br>

```python
abalone_df.mean()
```

```python
abalone_df.min()
```

```python
abalone_df.max()
```

<br>
- Plot histograms for numerical columns.<br>

```python
matplotlib.rcParams.update({'font.size': 40})
abalone_df.hist(figsize=(50, 40))
```

<br>
- Plot bar plots for categorical columns.<br>

```python
matplotlib.rcParams.update({'font.size': 100})
abalone_df['Sex'].value_counts().plot(kind='bar', figsize=(50, 40), rot=0).set(xlabel="Sex", ylabel="Occurrences", title="Male, Female, Infant Frequency (categorical)")
```
